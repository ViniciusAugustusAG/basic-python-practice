Faça um programa que recolhe a idade, gênero, cor dos olhos (azul, verde ou castanho) e cor dos cabelos (louros, castanhos ou pretos),
então determine quantas mulheres de olhos verdes e cabelos louros foram registradas.
O programa deve encerrar quando a idade digitada seja -1.

```python
mulheres = 0 # criamos nossas variáveis
idade = 0
maiorIdade = 0
j = 1
while idade !=-1:
    #Imprimir qual habitante está sendo registrado
    print("Habitande ",j)
    j += 1

    #Pedimos a idade do habitante
    idade = int(input('Idade do habitante (-1 para parar): '))
    if idade != -1: # se a idade é diferente de -1, o programa acontece
        if idade > maiorIdade: # A variável maioridade só atualiza o valor quando uma idade maior aparece
            maiorIdade = idade # maioridade recebe idade maior
        sexo = input('Sexo M (Masculino) ou F (Feminino): ') # pedimos o sexo do habitante
        cor_olhos = input('Olhos A (Azuis), V (Verdes) ou C (Castanhos): ') # recebemos cor dos olhos
        cor_cabelos = input('Cabelos L (Louros), C (Castanhos) e P (Pretos): ') # recebemos cor dos cabelos
        if (sexo in 'Femininofeminino') and (18 <= idade <= 35) and \
            (cor_olhos in 'Verdesverdes') and (cor_cabelos in 'Louroslouros'):
            # segregamos as mulheres que se encaixam no padrão
            mulheres = mulheres + 1 #contamos essas mulheres

print('A maior idade registrada é', maiorIdade,
      "\n Existem ",mulheres, 'mulheres de olhos verdes e louras.') #imprimimos resultado
```
