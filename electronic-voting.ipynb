```python
"""
Nome e número para os candidatos da questão:
Candidato João, número 1
Candidata Maria, número 2
Candidato Pedro, Número 3
Candidata Ana, número 4
"""
voto = -1

#Criar variáveis para contar os votos de cada possibilidade
nulos = 0
brancos = 0
joao = 0
maria = 0
pedro = 0
ana = 0

#conta o total de votos válidos feitos:
votos_totais = 0

while voto != 0:

    #pedir um voto para o usuário
    print('Urna eletrônica, digite o número de seu candidato e aperte enter:')
    voto = int(input("Digite:\n1-João\n2-Maria\n3-Pedro\n4-Ana\n"
                     "5-Nulo\n6-Branco\n0-Para encerrar a votação.\nSeu voto: "))

    #conta se o voto foi válido, se foi, adiciona no número de votos totais
    if  voto > 0 and voto < 7:
        votos_totais = votos_totais +1

    #verificar em quem o usuário votou
    if voto == 1:
        joao = joao +1
    elif voto == 2:
        maria = maria +1
    elif voto == 3:
        pedro = pedro +1
    elif voto == 4:
        ana = ana +1
    elif voto == 5:
        nulos = nulos +1
    elif voto == 6:
        brancos = brancos +1
    elif 0 > voto or voto > 6:
        print("Número inválido, digite novamente:")

    print("\n\n") 


#calcular porcentagem de votos nulos
porcentagem_n = 100 * nulos / votos_totais

#calcular porcentagem de votos brancos
porcentagem_b = 100 * brancos / votos_totais

#Imprimir na tela o que a questão pede
print(f"Contação dos votos:\nJoão - {joao} votos\nMaria - {maria} votos\nPedro - {pedro} votos\nAna - {ana} votos")
print(f"Votos nulos: {nulos} \nVotos em branco: {brancos}")
print(f"Porcentagem de votos nulos sobre o total de votos: {porcentagem_n:.2f}%")
print(f"Porcentagem de votos nulos sobre o total de votos: {porcentagem_b:.2f}%")
```
