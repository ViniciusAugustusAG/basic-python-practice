```python
'''
Crie um algoritmo que simule o funcionamento de um caixa de supermercado.
O caixa fica aberto até o fim do expediente e pode processar a compra de vários 
clientes. Cada cliente pode comprar vários itens. Ao ler cada item deve ser 
exibida uma mensagem para o operador do caixa perguntando se há mais itens a
serem processados. Ao final, exiba quanto a compra custou ao cliente. E então 
solicite do operador do caixa a informação se deseja fechar o caixa. Quando o 
caixa for fechado, exiba quanto de dinheiro aquele caixa apurou no dia.
'''

#Passo 1 -  Achar as variáveis
#Passo 2 - Achar o que fazer com as variáveis
#passo 3 - Dar um retorno para o usuário
```

```python

# variáveis iniciais:
nome_item = ""
preco_item = 0
lucro = 0

# primeiro while = dia de trabalho do caixa
j = 0
while j == 0:
    compra_cliente = 0 #zeramos o valor final da compra a cada loop
    i = 0 #zeramos o i a cada loop, pra podermos fazer um nova compra

    # segundo while = compra do cliente
    while i == 0:
        nome_item = input("Digite o nome do item: ")
        preco_item = float(input("Digite o preço do item: "))
        compra_cliente = compra_cliente + preco_item # somamos o novo produto aos outros já comprados
        i = int(input("Deseja adicionar outro item? \n" # \n quebra a linha, como um enter
                     "Sim (0), Não (1): "))

    lucro = lucro + compra_cliente # somamos o valor da nova compra no lucro que fizemos até então
    j = int(input("Deseja fechar o caixa?\n"
                  "Sim (1), Não (0): "))
    print("\n") # pula uma linha entre cada compra, pra ficar visualmente mais agradável

print(f"O caixa arrecadou: {lucro}") #print(f"texto {variável}") Usamos o 'f' antes das chaves e
                                        # colocamos {} com as variáveis que queremos imprimir dentro

```
