```python
'''Leia dois vetores de mesmo tamanho. Gere um terceiro
    vetor contendo os elementos que se repetem
    em ambos os vetores, na mesma posição.'''

tam = int(input('Tamanho do vetor: ')) #recebemos o tamanho dos vetores

vetor1 = [] # iniciamos nossos vetores
vetor2 = []
resultado = []

for i in range(2): # esse for faz 2 loops, cada loop preenche uma lista
    print(f"\nPreencha vetor {i+1}") # anunciamos o vetor atual

    for j in range(tam): # pedimos 1 elemento a cada loop, preenchendo o vetor
        if i == 0: # se etamos no primeiro loop i igual a 0, preenchemos o vetor 1
            aux = int(input(f"Digite elemento {j+1}: ")) # pedimos um elemento
            vetor1.append(aux) # adicionamos o elemento à nossa lista
        else: # se não estamos no primeiro loop preenchemos o segundo vetor
            aux = int(input(f"Digite elemento {j+1}: ")) # pedimos um elemento
            vetor2.append(aux) # preenchemos o nosso vetor 2

    if i == 0: # imprimimos o vetor 1 se estamos no primeiro loop, i igual a 0
        print(f"Vetor preenchido: {vetor1}")
    else: # imprimimos vetor 2 se estamos no segundo loop
        print(f"Vetor preenchido: {vetor2}")

for i in range(tam): # comparamos as listas, elemento por elemento
    if vetor1[i] == vetor2[i]: # se os elementos são iguais e estão na mesma posição
        resultado.append(vetor1[i]) # adicionamos o elemento ao vetor resultado

print(f"Os elementos em comum e na mesma posição são: {resultado}\n")
```
