Calcule a média das idades e defina se um grupo de pessoas é uma turma jovem, adulta ou idosa.
Considere que um adulto é uma pessoa entre 25 e 60 anos.
Imprima a média das idades, a idade mais alta e a mais baixa.

```python
#soma das idades
idade_total = 0

#idade de um indivíduo
idade = 0

#Quantidade de pessoas a ser cadastradas
n = int(input("Digite o número de pessoas: "))

#idade mais alta
mais_alta = 0
#idade mais baixa
mais_baixa = 0
#auxiliar para contagem
aux= 0


for i in range(n):
    print("USUÁRIO ", i+1)

    idade = int(input("Digite sua idade: "))
    print("\n\n") #pula 2 linhas

    idade_total = idade_total + idade
    if idade > aux: mais_alta = idade
    elif idade < aux: mais_baixa = idade

    aux = idade

media = idade_total/n

print(f"A idade mais alta é {mais_alta}. \n"
    f"A idade mais baixa é {mais_baixa} \n"
    f"A média das idades é igual a {media}")

if media > -1 and media < 26 :
    print("A turma é jovem.")
elif media > 25 and media < 60:
    print("A turma é adulta.")
else: 
    print("A turma é idosa.")
```
