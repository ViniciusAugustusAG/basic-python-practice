Determine se um uma entrada categoriza um polígono e, se possível, calcule a área do polígono. 

```python
lado = int(input('Quantos lados tem o poligno: ')) # pegamos a quantidade de lados
medida = float(input('Medida dos lados em cm: ')) # pegamos medida dos lados

if lado == 3: # comparamos a quantidade de lados com 3
  print('é um triângulo e sua área é {:.2f}'.format((medida**2*3**(1/2))/4))
elif lado == 4: # comparamos a quantidade de lados com 4
  print('é um quadrado e sua área é {:.1f}'.format((medida**2)))
elif lado == 5: # comparamos a quantidade de lados com 5
  print('é um pentángono')
elif lado < 3: # descobrimos se a quantidade de lados é menor que 3
  print('não é um polígono')
else: # Se nenhuma das comparações anteriores é verdadeira
  print('polígono não identificado')
```
